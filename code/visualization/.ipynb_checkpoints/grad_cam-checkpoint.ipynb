```python
def load_image(file_path):
    return pydicom.dcmread(file_path)

def show_image(img, title=''):
    plt.title(title)
    plt.imshow(img.pixel_array, cmap=plt.cm.bone)
    
def show_grayscale_image(img, title=''):
    plt.title(title)
    plt.imshow(img.pixel_array, cmap=plt.cm.gray, vmin=0, vmax=1)
```

```python
img0 = load_image(img_file_0)
show_image(img0, "Class 0")
```

```python
img1 = load_image(img_file_1)
show_image(img1, "Class 1")
```

```python
img2 = load_image(img_file_2)
show_image(img2, "Class 2")
```

```python
img3 = load_image(img_file_3)
show_image(img3, "Class 3")
```

## Load trained model and predict

```python
# img = img0.pixel_array
# height, width = img.shape
```

```python
# model checkpoint paths
ckpt_dir = '/data/vision/polina/projects/nn-visualization/data/models2'
ckpt_meta = os.path.join(ckpt_dir, 'model.ckpt-3948.meta')
ckpt_file = os.path.join(ckpt_dir, 'model.ckpt-3948')
```

```python
height = 2048
width = 2048
binary = '0v1v2v3'
test_img_dir = '/data/vision/polina/projects/chestxray/work_space/labels_after_segmentation_removingLineBreaks/classification/new_sub_v2/resized_'+str(height)+'_'+str(width)+'/'+binary+'/test_resized/'
img_path = os.path.join(test_img_dir, '1.npy')
img = np.expand_dims(np.load(img_path), 2)
```

```python
# restore the checkpoint
sess = tf.Session()
saver = tf.train.import_meta_graph(ckpt_meta)
saver.restore(sess, ckpt_file)

# access and create placeholders
graph = tf.get_default_graph()
x = sess.graph.get_tensor_by_name('input_node:0')
out = sess.graph.get_tensor_by_name('output_node:0')
```

```python
# predict
output = sess.run(out, feed_dict={x: [img]})
print("output: " + str(output))
```
